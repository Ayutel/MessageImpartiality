```python
frame = []
def make_frame(file):
    f = open(file, 'rb').read().decode('UTF-8').lower()
    h = f.split("&&")
    print(len(h))
    for i in range(0, len(h)):
        t = h[i].split('***')
        try:
            if(i==0):
                curr= {"tweet": t[len(t)-1].split('~~')[0], "date":t[len(t)-1].split('~~')[1], "tweet_id":t[0]}
            else:
                curr= {"tweet": t[len(t)-1].split('~~')[0], "date":t[len(t)-1].split('~~')[1], "tweet_id":t[0].split('\n')[2]}
        except:
            print(h[i])
        frame.append(curr) 
    return frame

data = make_frame("tweets.txt")
len(data)
```
