```python
import nltk
nltk.download("punkt")
```

```python
import re, pprint
from nltk import word_tokenize
from nltk.corpus import stopwords

def make_dict(file):
    f = open(file,'rb').read().decode('UTF-8').lower()
    tokens = word_tokenize(f)
    stop = set(stopwords.words('english'))
    stop.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}','#','@','rt','&','!','~',  "..."])
    tokens = [i for i in tokens if i not in stop]
    return tokens

vocabulary = make_dict("onlytext.txt")
```

```python
import numpy as np
tweets  = df["tweet"].as_matrix()
tweet_vectors = []
for line in tweets:
    word_vec = np.zeros(len(vocabulary))
    words = re.split(" ", line)
    for word in words:
        if word in vocabulary:
            word_vec[vocabulary.index(word)] = 1
    tweet_vectors.append(word_vec)
```

```python
republicans = df[df["affiliation"]==2].index.tolist()
democrats = df[df["affiliation"] == 1].index.tolist()
```

```python
rep_words = np.zeros(len(vocabulary))
for each in republicans:
    rep_words += tweet_vectors[each]
demo_words = np.zeros(len(vocabulary))
for each in democrats:
    demo_words += tweet_vectors[each]
```

```python
from math import log
word_bias = np.zeros(len(vocabulary))
total = len(democrats) + len(republicans)
for i in range(0, len(vocabulary)):
    if demo_words[i] !=0 and rep_words[i]!=0:
        word_bias[i] = max((demo_words[i]/total)*log(demo_words[i]/rep_words[i]), (rep_words[i]/total)*log(rep_words[i]/demo_words[i]))
    elif demo_words[i]==0 and rep_words[i]==0:
        word_bias[i] = 0
    else:
        word_bias[i] = 1
```

```python
tweet = "Will there be a government shutdown over the battle to #DefundObamaCare?"
```

```python
max_bias = 0
for word in tweet.split(" "):
    if word in vocabulary:
        bias = word_bias[vocabulary.index(word)]
        if bias > max_bias:
            max_bias = bias
print(max_bias)
```
